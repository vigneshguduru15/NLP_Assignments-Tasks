```python
from transformers import pipeline

# Load chatbot model
chatbot = pipeline(
    "text-generation",
    model="gpt2"
)

print("Chatbot: Hello! I am your AI assistant.")

# Sample questions
questions = [
    "What is NLP?",
    "What is Python?",
    "What is Machine Learning?"
]

# Generate responses
for q in questions:

    print("\nYou:", q)

    response = chatbot(
        q,
        max_new_tokens=20,
        truncation=True,
        pad_token_id=50256
    )

    answer = response[0]['generated_text']

    print("Chatbot:", answer)
```
